```python
df = pd.read_csv(path_file, sep =';', encoding='windows-1251' )
```

```python
# - вывести в консоль список всех регионов (только уникальные значения, т.е. например, можно использовать множество)
for i in df['Регион'].unique().tolist():
    print(i)
```

```python
df['Город']  = df['Город'].astype('string')
```

```python
df.dtypes
```

```python
# - создать csv-файл, где будут храниться все города с дефисом '-' в названии города
df_city = df.loc[df['Город'].str.contains('-')]
df_city.to_csv('cities_with-.csv')
```

```python
# - создать N csv-файлов с данными по каждому региону (например, "temp\Оренбургская область.csv") и в котором будет отсутствовать столбец регион
regions = df['Регион'].unique().tolist()
for region in regions:
    df_region = df.loc[(df['Регион']==region)]
    df_region = df_region[['Город','Федеральный округ','lat', 'lng']]
    df_region.to_csv(f'temp\{region}.csv')
```

```python
# - вывести в консоль список городов, в названии которых есть буква 'ы'
list_cities = df.loc[df['Город'].str.contains('ы')]['Город'].to_list()
```

```python
list_cities
```

```python
# - вывести в консоль список городов, название которых содержит до 4-х букв включительно и отсортированные в обратном алфавитном порядке
list_cities = df['Город'].to_list()
```

```python
sorted([city for city in list_cities if len(city)>=4], reverse=True)
```

```python
# - вывести в консоль список регионов и количество городов в каждом регионе
regions = df['Регион'].unique().tolist()
for region in regions:
    df_region = df.loc[(df['Регион']==region)]
    print("Регион-",region, "Кол-во городов-",len(df_region))
```

```python
# - вывести в консоль список федеральных округов с количеством регионов и количеством городов
counties = df['Федеральный округ'].unique().tolist()
for county in counties:
    df_county = df.loc[(df['Федеральный округ']==county)]
    
    print("Федеральный округ-",county,'Кол-во регионов-',len(df_county['Регион'].unique().tolist()), "Кол-во городов-",len(df_county))
```

```python
# - вывести в консоль буквы алфавита по убыванию и рядом количество городов, начинающихся на каждую букву
cities = df['Город'].unique().tolist()
alf= "абвгдеёжзийклмнопрстуфхцчшщъыьэюя"
for letter in alf:
    count_letter =0
    for city in cities:
        if city[0].lower()==letter:
            count_letter+=1
    print(letter, '-', count_letter)
```

```python
# - вывести в консоль список ТОП-10 городов с самыми длинными названиями
cities = df['Город'].unique().tolist()
sorted([city for city in cities], key=lambda i:len(i),reverse=True)[0:10]
```

```python
# - вывести в консоль список ТОП-10 городов, в названии которых больше всего согласных букв
import re

cities = df['Город'].unique().tolist()
sorted([city for city in cities], key=lambda i:len(re.findall(r'[ауоыиэяюёе]', i)),reverse=True)[0:10]
```

```python
# - вывести в консоль список ТОП-10 букв, которые используются чаще всего в названии городов
alf= "абвгдеёжзийклмнопрстуфхцчшщъыьэюя"
list_letter =[]
for letter in alf:
    count_letter =0
    for city in cities:
        count_letter+=city.count(letter)
    list_letter.append([letter, count_letter])
```

```python
sorted(list_letter, reverse=True, key=lambda i: i[-1])[:10]
```

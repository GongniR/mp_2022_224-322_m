# Queue

```python
import queue
import random
import time

first = time.time()

q = queue.PriorityQueue()
for i in range(1000):
    q.put(random.randint(0, 100))
for i in range(1000):
    print(q.get(), end=' ')

print()
print("Прошло времени: ", time.time() - first, "секунд")
```

# List

```python
first = time.time()

q = []
for i in range(1000):
    q.append(random.randint(0, 100))
for i in range(1000):
    print(q[i], end=' ')

print()
print("Прошло времени: ", time.time() - first, "секунд")
```

# Deque

```python
import collections

first = time.time()

q = collections.deque()
for i in range(1000):
    q.append(random.randint(0, 100))
for i in range(1000):
    print(q[i], end=' ')

print()
print("Прошло времени: ", time.time() - first, "секунд")
```

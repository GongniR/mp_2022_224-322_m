```python
list_names = df.loc[df['Name'].str.contains('W')]['Name'].to_list()
list_names
```

```python
sorted([name for name in list_names if len(name) >= 15], reverse=True)
```

```python
countries = df['Nationality'].unique().tolist()
for country in countries:
    df_country = df.loc[(df['Nationality'] == country)]
    print("Страна:", country, '| Кол-во игроков:', len(df_country['Name'].unique().tolist()), "| Кол-во клубов:", len(df_country['Club'].unique().tolist()))
```
